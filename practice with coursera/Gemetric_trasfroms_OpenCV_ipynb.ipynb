# Scaling

Tạo ra hình ảnh

```python
image = np.zeros((6,6))
image[1:5, 1:5] = 255
image[2:4, 2:4] = 0
```

Hiển thị hình ảnh với hiển thị màu sắc bằng hàm **cmap**

```python
plt.imshow(image, cmap = 'gray')
```

```python
new_image = cv2.resize(image, None, fx =2, fy = 1, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(new_image, cmap = 'gray')
```

# Scale hình ảnh được tải lên

Tải 1 hình ảnh khác lên

```python
image_s = cv2.imread("sm.jpg")
```

```python
plt.imshow(image_s)
```

```python
plt.imshow(cv2.cvtColor(image_s, cv2.COLOR_BGR2RGB))
```

scale hình ảnh vừa tải với chiều rộng gấp đôi, giữ nguyên chiều cao

```python
new_image_s = cv2.resize(image_s, None, fx=2, fy = 1, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(cv2.cvtColor(new_image_s, cv2.COLOR_BGR2RGB))
```

Scale hình ảnh với chiều cao gấp đôi, chiều rộng giữ nguyên


```python
image_1 = cv2.resize(image_s, None, fx = 1, fy = 2, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
```

Scale hình ảnh với cả chiều rộng và chiều cao đều tăng gấp đôi

```python
image_2 = cv2.resize(image_s, None, fx = 2, fy = 2, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
```

Scale hình ảnh giữ nguyên chiều rộng và giảm chiều cao xuống phân nửa

```python
image_3 = cv2.resize(image_s, None, fx = 1, fy = 0.5, interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(cv2.cvtColor(image_3, cv2.COLOR_BGR2RGB))
```

# Translation

```python
tx = 100
ty = 0
```

```python
M = np.float32([[1,0,tx], [0,1,ty]])
```

```python
 columns, rows, _ = image_s.shape
```

Chúng ta sử dụng hàm **warpAffine** từ module cv2. Tham số đầu vào đầu tiên là một mảng hình ảnh, tham số đầu vào thứ hai là ma trận biến đổi M, và tham số cuối cùng là kích thước chiều dài và chiều rộng của hình ảnh đầu ra (cols, rows):

```python
new_image = cv2.warpAffine(image_s, M, (columns, rows))
```

```python
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
```

Nhận xét: Hình ảnh đã bị mất một khúc bên phải

```python
new_image_1 = cv2.warpAffine(image_s, M, (columns + tx, rows + ty))
```

```python
plt.imshow(cv2.cvtColor(new_image_1, cv2.COLOR_BGR2RGB))
```

```python
tx = 0
ty = 50
```

```python
M2 = np.float32([[1,0,tx], [0,1,ty]])
```

```python
new_image_2 = cv2.warpAffine(image_s, M2, (columns + tx, rows + ty))
```

```python
plt.imshow(cv2.cvtColor(new_image_2, cv2.COLOR_BGR2RGB))
```

# Rotation

Chọn góc xoay là theta, theo đơn vị là độ.

```python
theta = 45
```

```python
M3 = cv2.getRotationMatrix2D(center = (columns//2 -1, rows//2 -1), angle = theta, scale = 1)
```

```python
new_image_4 = cv2.warpAffine(image_s, M3, (columns, rows))
```

```python
plt.imshow(cv2.cvtColor(new_image_4, cv2.COLOR_BGR2RGB))
```
