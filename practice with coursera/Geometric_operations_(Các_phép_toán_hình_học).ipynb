```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open("3.jpg")
```

```python
image.show(title = "a happy girl")
```

```python
plt.imshow(image)
```

# Thay đổi kích thước hình ảnh (scale)

```python
width, height = image.size
```

```python
print(f"Chiều rộng là: {width} và chiều dài là {height}")
```

Lấy chiều rộng và chiều cao bằng hàm **.size**

```python
new_width = width * 2 # tăng chiều rộng lên gấp 2 lần
new_height = height   # vẫn giữ nguyên kích thước của chiều cao
```

```python
new_image = image.resize((new_width, new_height))
```

Thay đổi hình ảnh với các tham số kích thước mới new_width, new_height bằng hàm **.resize()**

```python
plt.imshow(image) # show ra ảnh gốc
```

```python
plt.imshow(new_image) # show ra ảnh sau khi đã scale theo chiều rộng (width)
```

```python
new_width_2 = width   # giữ nguyên chiều rộng
new_height_2 = height * 2 # tăng chiều dài lên gấp đôi
```

```python
new_image_2 = image.resize((new_width_2, new_height_2))
```

```python
plt.imshow(new_image_2) # show ảnh đã scale theo chiều cao (height)
```

```python
new_width_3 = width * 2    # scale cả kích thước chiều rộng và chiều cao
new_height_3 = height * 2
```

```python
new_image_3 = image.resize((new_width_3, new_height_3))
```

```python
plt.imshow(new_image_3) # show ảnh đã scale theo cả chiều rộng và chiều cao
```

```python
new_width_4 = width // 2 # scale chiều rộng ảnh xuống còn phân nửa
new_height_4 = height // 2 # scale chiều cao ảnh xuống còn phân nửa
```

```python
new_image_4 = image.resize((new_width_4, new_height_4))
```

```python
plt.imshow(new_image_4) # hiển thị ảnh sau khi đã co lại chiều rộng và chiều cao đều giảm xuống còn một nửa
```

# Xoay ảnh

```python
theta = 45
```

khởi tạo tham số theta = 45 độ để xoay ảnh

```python
new_image_5 = image.rotate(theta)
```

thực hiện xoay ảnh với hàm **rotate** với tham số đầu vào là góc xoay **theta**

```python
plt.imshow(new_image_5) # hiển thị ảnh sau khi xoay 1 góc 45 độ
```

```python
new_image_6 = image.rotate(90) # thử xoay ảnh với góc 90 độ
```

```python
plt.imshow(new_image_6) # hiển thị ảnh sau khi đã xoay 1 góc 90 độ
```

```python
new_image_7 = image.rotate(180) # xoay ảnh với góc 180 độ
```

```python
plt.imshow(new_image_7) # hiển thị ảnh với gốc xoay là 180 độ
```
